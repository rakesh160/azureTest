```
HZ_CUST_ACCOUNTS=spark.read.parquet(fileSystemUrl+"/datalake/raw_data/EBS/InitialLoad/HZ_CUST_ACCOUNTS/")
HZ_PARTIES=spark.read.parquet(fileSystemUrl+"/datalake/raw_data/EBS/InitialLoad/HZ_PARTIES/")
```

```
# display(HZ_PARTIES)
```

```
HZ_CUST_ACCOUNTS.registerTempTable("HZ_CUST_ACCOUNTS")
HZ_PARTIES.registerTempTable("HZ_PARTIES")
```

```
CUST_ACC_D = spark.sql("""
SELECT 
INT(HZ_CUST_ACCOUNTS.CUST_ACCOUNT_ID) as INTEGRATION_ID,
date_format(HZ_CUST_ACCOUNTS.CREATION_DATE,"yyyy-MM-dd")                    CREATED_ON_DT,
  date_format(HZ_CUST_ACCOUNTS.LAST_UPDATE_DATE,"yyyy-MM-dd")                 CHANGED_ON_DT,
  INT(HZ_CUST_ACCOUNTS.CREATED_BY)                      CREATED_BY_ID,
  INT(HZ_CUST_ACCOUNTS.LAST_UPDATED_BY)                 CHANGED_BY_ID,
  CURRENT_DATE                                              INSERT_DT,    
  CURRENT_DATE                                              UPDATE_DT, 
  'EBS'                                                     DATASOURCE_NUM_ID,
  date_format(current_timestamp(),'yyyyMMddhhmmssSSS')   as LOAD_BATCH_ID,
 INT(HZ_CUST_ACCOUNTS.CUST_ACCOUNT_ID)        CUST_ACCOUNT_ID, 
    INT(HZ_CUST_ACCOUNTS.PARTY_ID)               PARTY_ID, 
    HZ_CUST_ACCOUNTS.ACCOUNT_NUMBER         ACCOUNT_NUMBER, 
    HZ_CUST_ACCOUNTS.ACCOUNT_NAME           ACCOUNT_NAME,   
CASE WHEN HZ_CUST_ACCOUNTS.CUSTOMER_TYPE = 'I' then 'Internal'
     WHEN HZ_CUST_ACCOUNTS.CUSTOMER_TYPE = 'R' then 'External'
     ELSE 'Unspecified'
END CUSTOMER_TYPE,
HZ_PARTIES.PARTY_NUMBER                 REGISTRATION_ID,
'N'                                     DELETE_FLG
FROM
HZ_CUST_ACCOUNTS LEFT OUTER JOIN HZ_PARTIES
ON HZ_CUST_ACCOUNTS.PARTY_ID = HZ_PARTIES.PARTY_ID
WHERE HZ_PARTIES.PARTY_TYPE = 'ORGANIZATION' 

""")
```
